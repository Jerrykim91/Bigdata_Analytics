### 컬럼 설명
- TOWN : 지역이름  
- LON, LAT : 위도, 경도 정보  
- **CMEDV : 해당 지역의 집값(중간값)**  
- CRIM : 근방 범죄율  
- ZN : 주택지 비율  
- INDUS : 상업적 비지니스에 활용되지 않는 농지 면적  
- CHAS : 경계선에 강에 있는지 여부  
- NOX : 산화 실소 농도  
- RM : 자택당 평균 방 갯수  
- AGE : 1940년 이전에 건설된 비율  
- DIS : 5개의 보스터 고용 센터와의 거리에 따른 가중치 부여  
- RAD : radial 고속도로와의 접근성 지수  
- TAX : 10000달러당 재산세  
- PTRATIO : 지역별 학생 - 교사 비율  
- B : 지역의 흑인 지수 (1000(B - 0.63)^2) B는 흑인의 비율  
- LSTAT : 빈곤층의 비율

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler_columns = cols[1:]

df[scaler_columns] = scaler.fit_transform(df[scaler_columns])
# 민맥스는 가진 값안에서 진행 -> 회귀 분류 
# 스케터는 예측 
````

````python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
````

````python
X = df[scaler_columns]
y = df['CMEDV']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,random_state = 33)
````

````python
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
````

````python
print(lr.coef_)
````

````python
plt.rcParams['figure.figsize']=[12,16]
coefs = lr.coef_.tolist()
coefs_series = pd.Series(coefs)
````

````python
X_labels = scaler_columns
ax = coefs_series.plot.barh()
ax.set_title('feature_coef_graph')
ax.set_xlabel('coef')
ax.set_ylabel('x_features')
ax.set_yticklabels(X_labels)
plt.show()
````

# 모델 평가 

- R2 스코어 
- rmse 


## R2 스코어 

````python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
````

````python
# 데이터를 세세하게 보려면 굳이 루트가 필요 없다. 
````

````python
# rmse 
y_predictions = lr.predict(X_train)
print(mean_squared_error(y_train, y_predictions))
````

````python
# rmse
y_predictions = lr.predict(X_test)
print(mean_squared_error(y_test, y_predictions))
````

````python
# 크기가 큰경우 루트를 씌워서 보면 명확하게 데이터의 변화에 대해 알수있다. 
````

````python
# rmse + 루트 -> 
y_predictions = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions)))
````

````python
# rmse + 루트
y_predictions = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions)))
````

## f벨류 -> 정리가 필요 
-> 분산 동일한지
f는 왜보는가 ? 평균분산을 봐야하는 이유 -> 앞에 모듈이랑 현재 모듈간에 비교를 위해서 -> 차이가 난다면 앞의 모델과의 성질이 달라진것 -> 모델간의 데이터 성향이 변화된것을 설명 해준다. 
-> 성능 차이를 확인 가능하다는거 
-> 유의한가를 보는거 -> 많이 달라진다면 우연히 나온것 -> f에대한 피벨류를 같이 봐야한다. 
- 각각 독립변수간에 피벨류 

[F-value의 의미와 분산분석](https://angeloyeo.github.io/2020/02/29/ANOVA.html)

````python
# 다중 공선성 

from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()

vif['VIF Factor'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['features'] = X_train.columns
vif.round(1)
````
