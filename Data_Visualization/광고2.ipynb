### 중점관리키워드 추출
- 노출수, 클릭수 기준 상위 5%에 해당하는 키워드 추출  
 - 95 백분위수 찾기 (quantile 함수 사용)  
 - 95 백분위수 이상의 노출수 추출  
 - 상위 5%에 해당하는 키워드 추출

```python
df_index = df.set_index('키워드')
```

```python
df_index.head()
```

```python
imp = df_index['노출수']
imp = imp[imp >= imp.quantile(0.95)]
```

```python
imp.head()
```

```python
clk = df_index['클릭수']
clk = clk[clk >= clk.quantile(0.95)]
```

```python
clk.head()
```

```python
# 노출수와 클릭수 모두 95 백분위수 이상 인 키워드
imp = df_index['노출수']
clk = df_index['클릭수']
```

```python
result = df_index[(imp >= imp.quantile(0.95)) & (clk >= clk.quantile(0.95))]
```

```python
result.head()
```

```python
result.index
```

### 노출수, 클릭수 기준 상위 5%의 키워드가 아닌데도 비용이 많이 쓰이는 키워드
- 노출수 95백분위수 미만  
- 클릭수 95백분위수 미만  
- 총비용 85백분위수 이상  
- 총비용 95백분위수 미만  

```python
cost = df_index['총비용(VAT포함,원)']
```

```python
result = df_index[(imp < imp.quantile(0.95)) & (clk < clk.quantile(0.95))
                 & (cost >= cost.quantile(0.85)) & (cost < cost.quantile(0.95))]
```

```python
result.index
```

### 광고그룹 분석
- as-is : 키워드 기준 분석   
 - 키워드의 노출수/클릭수/총비용...  
- to-bo : 광고그룹 기준 분석  
 - 데이터의 구성을 광고그룹 기준으로 변경  
 - 광고그룹의 노출수/클리수/총비용...   
 - groupby 함수 사용   

```python
grouped = df.groupby('광고그룹')
```

```python
grouped.count()
```

```python
grouped.mean()
```

```python
grouped.median()
```

```python
grouped.sum()
```

```python
df_group = grouped.sum()
df_group
```

```python
df_group['클릭률(%)'] = df_group['클릭수'] / df_group['노출수']
```

```python
df_group['평균클릭비용(VAT포함,원)'] = df_group['총비용(VAT포함,원)'] / df_group['클릭수']
```

```python
df_group[df_group['클릭수'] == 0]
```

```python
df_group['평균클릭비용(VAT포함,원)'] = df_group['평균클릭비용(VAT포함,원)'].fillna(0)
```

```python
df_group[df_group['클릭수'] == 0]
```

```python
# 평균클릭비용 컬럼 원단위 반올림, int 타입 처리
df_group['평균클릭비용(VAT포함,원)'] = round(df_group['평균클릭비용(VAT포함,원)'], 0)
df_group['평균클릭비용(VAT포함,원)'] = df_group['평균클릭비용(VAT포함,원)'].astype(int)
```

```python
df_group.head()
```

### 중점관리 광과그룹 / 저효율 광고그룹
- 중점관리 광고그룹
 - 상위 20% (노출수 80백분위수 이상)
 - 상위 10% (클릭수 90백분위수 이상)

```python
imp = df_group['노출수']
clk = df_group['클릭수']
```

```python
result = df_group[(imp >= imp.quantile(0.8)) & (clk >= clk.quantile(0.9))]
```

```python
result.index
```

### 저효율 광고그룹 : 노출수, 클리수 기준 상위의 키워드가 아닌데도 비용이 많이 쓰이는 과그룹
- 노출수 80백분위수 미만  
- 클릭수 90백분위수 미만  
- 총비용 60백분위수 이상
- 총비용 90백분위수 미만

```python
cost = df_group['총비용(VAT포함,원)']
```

```python
result = df_group[(imp < imp.quantile(0.8)) & (clk < clk.quantile(0.9))
                 & (cost >= cost.quantile(0.6)) & (cost < cost.quantile(0.9))]
```

```python
result.index
```
