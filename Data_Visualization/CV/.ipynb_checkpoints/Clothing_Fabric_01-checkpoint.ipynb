````python
random.seed(0) # 시드값 부여 

defect_list = glob.glob(PATH_DEFECT+'*.png')
mask_list   = glob.glob(PATH_MASK +'*.png')
pass_list   = glob.glob(PATH_NODEFECT+'**/*.png')
````

````python
# pass_list
````

````python
for defect in defect_list:
    num = defect.split('/')[-1].split('_')[0]
    for mask in mask_list:
        num_mask = mask.split('/')[-1].split('_')[0]
#         print(num_mask)
        if num == num_mask:
            print(num)
````

````python
# 이름 추출 
for defect in defect_list:
    num = defect.split('\\')[-1].split('_')[0]
#     print(defect)
#     print(defect.split('/')[-1])
#     print(defect.split('/')[-1].split('_')[0])
#     print('='*30)
    for mask in mask_list:
#         print(mask.split('/')[-1])
#         print(mask.split('/')[-1].split('_')[0])
#         print('='*30)
        num_mask = mask.split('\\')[-1].split('_')[0]
        print(num_mask)
        if num != num_mask:
            print(num)
````

````python
# Match_defect-make pairs
new_defect_list = list()
new_mask_list   = list()

# 각자 이름 추출 
for defect in defect_list:
    num = defect.split('\\')[-1].split('_')[0]
#     print(len(num))
    for mask in mask_list:
        num_mask = mask.split('\\')[-1].split('_')[0]
#         print((len(num)),len(num_mask))
        # 이거 이해가 안되 
        if num == num_mask:
            new_defect_list.append(defect)
            new_mask_list.append(mask)
````

````python
defect_list = new_defect_list
mask_list   = new_mask_list
````

# 두번째 데이터 셋 구성 

````python
if os.path.exists('data/dataset/2') is False:
    os.mkdir('data/dataset/2')
if os.path.exists('data/dataset/2/OK') is False:
    os.mkdir('data/dataset/2/OK')
if os.path.exists('data/dataset/2/FAIL') is False:
    os.mkdir('data/dataset/2/FAIL')
````

````python
os.path.basename(os.getcwd())
````

````python
import os 

def func_dir(adr):
    base = os.getcwd()
    full = base + adr
    tmp = adr.split('\\')

    for idx in range(len(tmp)):
        ck = base +'\\'+'\\'.join(tmp[:idx+1])
        if os.path.exists(ck) is False:
            os.mkdir(ck)
            print('데이터 확인 :', ck , os.path.isdir(ck))
        else:
            print('이미 데이터 존재 -> PASS')

    return 'DONE'
````

````python
func_dir('data\\dataset\\2\\FAIL')
````

````python
tmp = ['data/dataset/2' ,'data/dataset/2/OK', 'data/dataset/2/FAIL' ] 

for file_name in tmp : 
#     print(file_name)
    func_dir(file_name)
    print(file_name, ' >>>>  완료')
````
