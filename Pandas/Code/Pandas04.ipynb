### 앤스콤 데이터 집합 불러오기 

```python
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
anscombe = sns.load_dataset("anscombe")
```

```python
print(anscombe)
print(type(anscombe))
```

```python
# %matplotlib inline
# import matplotlib.pyplot as plt
```

```python
dataset_1 = anscombe[anscombe['dataset']=='I']
```

```python
# 그래프 그리기
plt.plot(dataset_1['x'], dataset_1['y'])
```

```python
# 그래프 그리기 => 0 이아니라 o임
plt.plot(dataset_1['x'], dataset_1['y'],'o')
```

```python
# 불린 추출  => 각 변수에 담는다 
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
```

```python
# 기본 틀 만들기 
fig = plt.figure()
# 그래프 격자를 그림
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

fig.tight_layout()
```

```python
# 틀을 기반으로 그리기 , 점으로 
axes1.plot(dataset_1['x'], dataset_1['y'],'o')
axes2.plot(dataset_2['x'], dataset_2['y'],'o')
axes3.plot(dataset_3['x'], dataset_3['y'],'o')
axes4.plot(dataset_4['x'], dataset_4['y'],'o')

# 출력 
fig
```

```python
# 그래프이름 달기 = 타이틀달기  
axes1.set_title("daraset_1")
axes2.set_title("daraset_2")
axes3.set_title("daraset_3")
axes4.set_title("daraset_4")

# 출력 
fig
```

```python
# 대타이틀 
fig.suptitle("Anscombe Data")

# 출력 
fig
```

```python
# 위그래프처럼 레이아웃을 조절하기위한 함수  
fig.tight_layout()

fig
```

### 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
# 틀(fig) 만들기
fig = plt.figure()
# 두개는 한세트 
axes = fig.add_subplot(1,1,1)
```

```python
axes.hist(tips['total_bill'],bins=10)
axes.set_title('Histogram_Total Bill')
axes.set_xlabel('Frequency')
axes.set_ylabel('Total Bill')

#그래프로 출력 
fig
```

```python
scatter_plot = plt.figure()
axes = scatter_plot.add_subplot(1,1,1)
axes.scatter(tips['total_bill'],tips['tip'])
axes.set_xlabel('Total Bill')
axes.set_ylabel('Tip')
```

```python
boxplot = plt.figure()
axes = boxplot.add_subplot(1,1,1)

axes.boxplot(
    [tips[tips['sex']=='Female']['tip'],
    tips[tips['sex'] == 'Male']['tip']],
    labels = ['Female','Male'])

axes.set_xlabel('Sex')
axes.set_ylabel('Tip')
axes.set_title('Boxplot of Tips by Sex')
```
