```python
# 완성
def loadData( filename, dir='../input/ml-100k' ):
    
    data  = list()
    array = list()
    with open(f'{dir}/{filename}') as f :

        for line in f :
            uid, mid,rating,_ = line.split('\t')
            data.append({'uid':uid,'mid':mid})
            array.append(float(rating))
    
    array = np.array(array)
    return (data, array)

loadData('ua.base')
```
