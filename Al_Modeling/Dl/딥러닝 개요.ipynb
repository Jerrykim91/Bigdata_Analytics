# 딥러닝 개요

## 개요
---

- AI > ML > Deep Learning(딥러닝)
- 인공지능 연구 분야중 하나이다.
- 딥러닝은 인간의 뇌속의 뉴런을 모방한 신경망을 여러층으로 겹쳐서(다양하게 구성), 그 신경망을 대규로모 만들어 학습 및 예측을 수행하는 머신러닝의 한가지 방법이다.

- 3단게로 구성 
    - 망짜고 학습하고 예측하는것 
    
    - 1단계 : 모델 작성 
        - 신경망 설계, 구현 
        - 요구사항 검토 
            - 프로젝트 주제 등등..
            - 어떤 신경망형태로 구현 해야 적합한지 사전에 인지해야한다.
            - 새롭게 적용하는 것도 검토해야하며, 새로운 망 구성 검토해야한다.
        - 뉴럴네트워크 구조를 설계(작성)
        - 수치 계산(산수수준) 
            - 입력 레이어(층)수, 출력층의 수, 히든레이어의 수, 레이어의 종류등등, 수치를 정립 
        - 가중치 파라미터, 바이어스(편향값)들을 0 혹은 특정값으로 혹은 난수로 초기화 
            - 퍼셉트론을 통해서 등장 
      
    - 2단계 : 학습 
        - 입력데이터에 맞춰 적절한 예측값이 나오도록, 가중치 파라미터(W), 바이어스(b)를 최적화한다.
        - 이를 위해서 대량의 학습데이터와 정답 세트를 이용한다
            - 딥러닝은 기본적으로 지도학습의 개념을 가짐  <- 생성 모델 
            - 예측값으로 부터 최초학습데이터까지 거꾸로 이동 
     
    - 3단계 : 예측(추론)
        - 태스트 데이터를 모델에 입력하여서 예측을 수행, 이에 따른 판단

## 엔진의 구동 방식(양대 진영)
---

- 2019 집계(미국에서 조사가능한 기업 대상)
    - 1위 : 텐서플로우
    - 2위 : 파이토치
    - 3위 : 케라스 

- 방식 
    - Define-and-Run
        
        - 2016년도까지 주류의 스타일, 방식
            - 텐서플로우!!
                - 플로우 : 레이어를 만드는 방식

                1. 계산 그래프등 데이터 플로우 그래프를 설계     
                2. 세션에 그래프에 들어갈 데이터(수치)를 주입   
                3. 세션실행(연산, 학습, 훈련등등)
                4. 플로우(파이썬), 연산(C++), 분리
                5. 고속화 처리가 쉽게 구현된다
                6. 상대적인 학습비용이 발생, 높다
                7. 오류가 나면 도대체 어디가 잘못된 것인지 파악하기 어렵다.(에러 포인트를 찾기어렵다)

    - Define-by-Run
        - 채이너(chainer)에서 파이토치(pyTorch)분사 
        - 채이너에서 사용하는 방식(일본에서 많이 사용)
        - pyTorch 
            1. 계산 그래프를 그리는 동시에
            2. 주입할 데이터가(수치)를 주입, 보관
            3. 연산을 바로 수행 <- 파이썬 방식과 동일 
            4. 오류가 발생하면 즉시 파악하기 쉽다.
            5. 유연성, 직관성, 고성능 3 요소 - 강조
            6. 텐서플로우도 Eager Execution을 이용하면 이방식으로 처라가 가능하다.

## RNN
--- 

- 순환(재귀적) 뉴럴 네트워크
    - 시간정보가 들어간다

- 자연어 처리 
    - 문서 분류, 대화문 생성, 기계 번역, 문서 요약, 챗봇
        - 자연어 처리 프로세스
            - 사전 생성, 형태소 분리(분절), 데이터의 벡터화(원핫 인코딩)
                - 수치화(정형화) 해야 RNN으로 갈수있다. 

- 동영상 분류
    - http://tcpschool.com/deeplearning/deep_sound

- 음성인식 
    - project magenta (음성 합성)
        - https://github.com/tensorflow/magenta

- 음성 합성, 음악 생성(조합)
    - one - to - many
        - 하나의 입력에서 여러개가 출력
        - 이미지 캡셔닝, 이미지를 보고 텍스트를 생성

    - many - to - one
        - 여러개의 입력에서 출력 
        - 입력 문서(리뷰)가 긍정? 부정? 판별:감성분류
        - 스펨 메일 분류

    - many - to - many
        - 여러개의 입력에서 여러개의 출력
        - 입력 문장에 대해서 대답 문자을 만든다 -> 챗봇
        - 입력 문자으로부터 번역된 문자을 만든다 -> 기계번역
        - 입력대비 동시에 출력되는 방식

- 시계열 데이터 (설비, 품질, 제조, 주식, 금융등등)
    - LSTM 등등
