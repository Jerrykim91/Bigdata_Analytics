# 정의

- 머신러닝 > 강화학습
- Reinforcement Learning
- Agent(에이전트)가 Environment(환경)의 상태에 맞춰서 어떻게 Action(행동)을 수행해야, 가장 큰 reward(보상)을 받는지 찾아내는 방법
- 학습데이터가 없다
- 시행착오만으로, 경험으로 학습한다

# case

- 설정 : A씨는 요트를 타고 무동력 세계 일주를 진행하다가, 남극 인근에서 난파되어서, 무인도에 도착하였다. => 생존!!
- 구성요소를 정의
- Agent : A씨
- Environment : 무인도
- Action : 걷는다, 잔다, 밥먹는다, 사냥한다, 요리한다...
- Reward : 물, 동물, 식물(식량획득), 아이템 획득, 체력 획득 => 생존문제와 연결

# 구성 요소

## Agent (에이전트)
- 표류한 A씨

## Action (행동)

- 에이전트가 환경에서 일으키는 움직임
  - 상태(state) : 행동에 따라 변화하는 환경적인 요소

## Reward (보상)

- 동일한 행동을 취해도 상태에 따라 결과가 달라진다
- 보상은 생존에 귀결된다, 좋은 보상, 나쁜 보상

## Policy (정책)

- 현재 상태에 따라, 다음 행동을 결정하는 기준
- 즉, 다음 행동을 결정하기 위한 전략

## 즉시 보상, 지연보상

- 즉시 보상
  - 행동 직후 즉시 적용되는(받게되는) 보상
    - 휴식 하면 바로 체력이 채워짐
- 지연 보상
  - 미래에 발생되는 보상
    - 무인도를 10시간 탐사했다 -> 음식을 발견했다 
- 즉시 보상만 탐하다가, 나중에 발생할 지연보상(더 큰보상)을 놓칠수 있다

##  interset (수익)

- 즉시보상 + 지연보상 => 강화학습이 목표
- 에이전트의 행동에 따라 결과가 수반된다
  - 에이전트에게 달렸다

## Value (가치)

- 수익이라는 미래의 결론, 에이전의 상태와 정책을 고정하여서, 조건부로 수익을 계산
- 이런 기준에서 가치를 가장 크게 만들어 낼수 있는 조건을 찾아낸다 => 이것은 곳 학습한것이다

# 종합

- 가치의 최대화 -> 수익의 최대화 -> 보상을 많이 받는 정책을 강화학습의 목표로 연결

# 학습 전략, 사이클

- 에이전트는 최초에 어떤 행동을 해야 할지 판단을 할수 없다
  - 선택 가능한 행동에서 랜덤으로 수행
- 에이전트가 행동의 댓가로 보상을 받으면, 어떤 행동을 했을때, 어떤 보상을 받았는지 기억 => 경험 
- 경험을 기반으로 정책을 계산
- 이과정을 반복하여, 게임 종료시까지 많은 보상을 받을 수있는 정책을 계산

## 마르코프 결정 과정

- 바로 직전에 기술한 내용을 기반으로 한 학습 사이클을 의미
- MDP (Markov Decision Process)
- 현재 상태에서 선택한 행동에 따라, 다음 상황이 확정되는 시스템

## 에피소드 ( Episode)
- 게임 종료시 까지 학습 분량

## 스텝(Step)

- 1회 행동 분량 => 1스텝

# 정책 계산 방법

## 정책 반복법(Policy Iteration)

- 성공시 선택한 행동이 중요하다
  - 행동을 많이 수행
    - 이를 통해서 정책을 계산(갱신)
      - 정책 경사법(알로리즘)

## 가치 반복법(Value Iteration)

- 다음 상태의 가치와, 현재 상태의 가치의 차이를 계산
  - 차이만큼 현재 상태의 가치를 늘리는 기법
    - 알고리즘
      - Sarsa
      - Q Learning
      - DQN
