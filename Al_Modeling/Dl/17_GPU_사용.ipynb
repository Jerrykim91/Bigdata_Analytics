```
# GPU 적용하여 사용하는 지점 => Session.run()
# 데이터 플로우 그래프 만드는 지점부터 적용해도 관계 없디. 
import time
```

```
# GPU 적용하여 사용하는 지점 => Session.run()
# 데이터 플로우 그래프 만드는 지점부터 적용해도 관계 없다
start = time.time()
sess = tf.Session()
with tf.device('/device:GPU:0'):   # GPU 연산
    a = tf.placeholder( tf.int32, [3]  )
    b = tf.constant( 2 )
    x_opration = a * b
    res = sess.run( x_opration, feed_dict={ a:[1,2,3] } )
    print( res, type(res) )
sess.close()
gpu_delay_time = time.time() - start
print("time :", gpu_delay_time)
```

```
start = time.time()
sess = tf.Session()
with tf.device('/device:CPU:0'):  # CPU 연산  
    a = tf.placeholder( tf.int32, [3]  )
    b = tf.constant( 2 )
    x_opration = a * b
    res = sess.run( x_opration, feed_dict={ a:[1,2,3] } )
    print( res, type(res) )
sess.close()
cpu_delay_time = time.time() - start
print("time :", cpu_delay_time)
```

```
#from tensorflow import keras
#from tensorflow.keras import backend as K
from keras import backend as K
```

```
K.tensorflow_backend._get_available_gpus()
```

```
# 케라스 GPU 사용은 텐서플로우와 연계 작업
config  = tf.ConfigProto()
# 학습간 많이 만나는 오류가 => 메모리 부족 OOM( Out of Memory) 
# -> batch사이즈 조정, 데이터 크기 조정, 혹은 그래픽교체
# 학습간에 메모리가 부족하면 자동으로 증가시켜준다
# 학습전에 GPU 사용 메로리양을 제한할수 있다
config.gpu_options.allow_growth = True
sess    = tf.Session(config=config)
# 케라스에 세션을 연결해라
K.set_session( sess )

# 실전코드
with tf.device('/device:GPU:0'):
    # 케라스코드 : 모델링, 학습이후~
    pass

sess.close()
```
